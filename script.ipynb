```python
import requests
import json

# API url example
"""
https://xmart-api-public.who.int/WASHMART/GLAAS_EN_2
?$filter=Published
%20eq%20true
%20and%20Ind%20in%20
(%27COORD_FREQ%27,%27COORD_MECH%27,%27COORD_INGOV%27,%27COORD_INNGO%27,%27COORD_INDON%27,%27COORD_REVIEW%27,%27COORD_APPEV%27,%27COORD_FRAME%27,%27COORD_INDOC%27,%27COORD_BUDG%27,%27COORD_SECTORS%27)
%20and%20Time%20in%20
(%272013%27,%272013%27,%272016%27,%272016%27,%272018%27,%272018%27,%272021%27,%272021%27)
&$select=Ind,IndText,IndSeq,LocText,LocTypeText,Loc,Time,TimeTypeText,DataType,Dim1ValText,Dim1TypeText,Dim1Type,Dim1Val,Dim2ValText,Dim2TypeText,Dim2Type,Dim2Val,Dim3ValText,Dim3TypeText,Dim3Type,Dim3Val,Dim4ValText,Dim4TypeText,Dim4Type,Dim4Val,Dim5ValText,Dim5TypeText,Dim5Type,Dim5Val,ValCode,ValText,ValCodeNum,ValSeq,UoM,DataType,Com1TypeText,Com1,Com2TypeText,Com2,Com3TypeText,Com3
&$orderby=Time%20desc,IndSeq%20asc,Dim1ValSeq%20asc,Dim2ValSeq%20asc
"""
# api_url = "https://xmart-api-public.who.int/WASHMART/GLAAS_EN_2"
# seems to only contain 2021 data
api_url = "https://xmart-api-public.who.int/WASHMART/GLAAS_EN_2"

response = requests.get(api_url)

if response.status_code == 200:
    new_data = response.json()

    try:
        with open("data.json", "r") as json_file:
            existing_data = json.load(json_file)
    except (FileNotFoundError, json.decoder.JSONDecodeError):
        existing_data = []

    existing_data.append(new_data)

    with open("data.json", "w") as json_file:
        json.dump(existing_data, json_file, indent=4)
        print("Data appended to data.json file.")
else:
    print("Failed to retrieve data from the API. Status code:", response.status_code)
```
